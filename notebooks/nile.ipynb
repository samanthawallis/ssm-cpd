```python
import numpy as np
import pandas as pd

def load_nile():
    from statsmodels.datasets import nile
    d = nile.load_pandas().data
    nile_series = pd.Series(d["volume"].values, index=d["year"].values, name="NileVolume")
    return nile_series

nile = load_nile()
print("Nile:", nile.shape, "| years:", nile.index.min(), "â†’", nile.index.max())
```

```python
import matplotlib.pyplot as plt

t = np.arange(len(nile))
y_obs = nile.values
plt.plot(t, y_obs)
```
