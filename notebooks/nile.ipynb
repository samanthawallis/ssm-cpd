```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# import scienceplots
# plt.style.use(['ieee', 'grid', 'notebook'])
from statsmodels.datasets import nile

import sys
sys.path.append('..')
from src.detect import detect_single_cp
```

```python
def load_nile():
    d = nile.load_pandas().data
    nile_series = pd.Series(d["volume"].values, index=d["year"].values, name="NileVolume")
    return nile_series

df = load_nile()
```

```python
t = np.arange(len(df))
y = df.values
output = detect_single_cp(y, dt=1)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df.index, df.values)
plt.axvline(df.index[output['location_k']], color='k', linestyle='--', label='Detected CP')
plt.axvline(1898, color='darkgray', linestyle=':', label='True CP')
plt.xlabel('Year')
plt.ylabel('Flows')
plt.title('Nile River Flows at Ashwan 1871-1970 with Changepoints')
plt.legend()
plt.show()
```
